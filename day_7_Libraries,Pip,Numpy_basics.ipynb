```python
#NumPy Arrays vs Python Lists
my_list = [1, 2, 3, 4, 5]

# NumPy array
import numpy as np
my_array = np.array([1, 2, 3, 4, 5])

print(type(my_list))   # <class 'list'>
print(type(my_array))  # <class 'numpy.ndarray'>
```

```python
#vectorization
import numpy as np
array1=np.array([1,3,4,5])
array2=np.array([3,4,5,6])
print(array1+array2)
print(array1*array2)
print(array1/array2)
```

```python
import numpy as np
scores=np.array([55,60,75,80,90]) # Convert to a NumPy array
print(scores+5)
print(scores*2)
print(scores/2)

print(np.min(scores))
print(np.max(scores))
print(np.mean(scores))
print(np.median(scores))
print(np.std(scores))
```

```python
import numpy as np

np.zeros(5)             # [0. 0. 0. 0. 0.]
np.ones(5)              # [1. 1. 1. 1. 1.]
np.arange(0, 10, 2)     # [0 2 4 6 8] — like range() but numpy
np.linspace(0, 1, 5)    # [0.   0.25 0.5  0.75 1. ] — 5 evenly spaced numbers
np.random.randint(1, 10, size=5)  # 5 random integers between 1 and 9
```

```python
import numpy as np
marks = np.array([55, 60, 75, 80, 90,100])

print("marks",marks)
print("Average",np.mean(marks))
print("Median",np.median(marks))
print("Standard Deviation",np.std(marks))
print("Above 80",marks[marks>80])
```

```python
import numpy as np

students = np.array([45, 67, 89, 23, 78, 95, 56, 88, 72, 61])

print("Mean:", np.mean(students))
print("Median:", np.median(students))
print("Std:", np.std(students))
print("Passed (above 50):", students[students > 50])
print("Total passed:", len(students[students > 50]))
```
